```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import pandas as pd
import numpy as np
```

```python
# plots a finite element mesh
def plot_fem_mesh(nodes_x, nodes_y, elements):
    # print(nodes_x)
    it = 0
    for element in elements:
        x = [nodes_x[element[0]], nodes_x[element[1]], nodes_x[element[2]]]
        y = [nodes_y[element[0]], nodes_y[element[1]], nodes_y[element[2]]]
        # print(x , y)
        plt.fill(x, y, edgecolor='black', fill=False)
        it += 1
        
```

```python
df_x_y = pd.read_csv("/home/san/Code/2d_Laplace_eq_FEM/meshes/mesh/mesh_nodes.csv")
df_z = pd.read_csv("/home/san/Code/2d_Laplace_eq_FEM/src/solution.csv")
df_elements = pd.read_csv("/home/san/Code/2d_Laplace_eq_FEM/meshes/mesh/mesh_polygons.csv")
for element in df_elements.values:
    element -= 1;

triangulation = tri.Triangulation(df_x_y["x"].values, df_x_y["y"].values, triangles=df_elements.values)
plt.tricontourf(triangulation, df_z["sol"].values)
# show
#plot_fem_mesh(df_x_y["x"].values, df_x_y["y"].values, df_elements.values)

plt.colorbar()
plt.axis('equal')
plt.show()
```
